# 2. Importación de ficheros
    Se especifica el directorio donde están los ficheros .csv en bruto.
    Se crea una lista con los nombres de los ficheros .csv en bruto.
    Se recorre esta lista importando cada uno de estos ficheros como DATAFRAME.
    Finalmente se obtiene una lista de DATAFRAMES con los dataframes ya importados.
    Se testea que la importación se ha realizado correctamente.

## Static
Importación de los ficheros en bruto estáticos

```python
raw_static_path = "../data/raw/static"
raw_static_name = os.listdir(raw_static_path)
raw_static = []
for i in raw_static_name:
    aux = pd.read_csv("../data/raw/static/" + i )
    raw_static.append(aux)

if len(raw_static_name) == len(raw_static):
    print("Import static correct")
    print("Number of dataframes:" + str(len(raw_static)))  
else:
    print("Import incorrect")
```

## Dynamic

```python
# TRAIN
raw_train_path = "../data/raw/dynamic/train"
raw_train_name = os.listdir(raw_train_path)
raw_train = []
for i in raw_train_name:
    aux = pd.read_csv("../data/raw/dynamic/train/" + i )
    raw_train.append(aux)

if len(raw_train_name) == len(raw_train):
    print("Import train correct")
    print("Number of dataframes:" + str(len(raw_train)))
else:
    print("Import incorrect")
```

# Duplicados

## Static

```python
fig, ax = plt.subplots(figsize=(10,3))
dupl_list = []
df_list = []
count = 1
sum = 0
for aux in raw_static:
    dupl = aux.duplicated().sum()
    dupl_list.append(dupl)
    df_list.append(count)
    count+=1
    sum = sum + dupl

ax.bar(df_list,dupl_list)

#Details
ax.set_title("Distribution of duplicates in Static")
ax.set_xlabel("Dataframe")
ax.set_ylabel("Nº of Duplicates Rows")
fig.savefig("../Figures/Static_duplicates.png", bbox_inches='tight')
print("El número de valores duplicados en los Dataframes Static TOTAL es: " + str(sum))
```

```python
sum = 0
for aux in raw_static:
    aux.drop_duplicates(subset = "Timestamp", keep='first', ignore_index=True, inplace=True)
    aux.index.name = "mseg"
print("Duplicados borrados con exito")
```

```python
fig, ax = plt.subplots(figsize=(10,3))
dupl_list = []
df_list = []
count = 1
sum = 0
for aux in raw_static:
    dupl = aux.duplicated().sum()
    dupl_list.append(dupl)
    df_list.append(count)
    count+=1
    sum = sum + dupl

#Graph
ax.bar(df_list,dupl_list)

print("El número de valores duplicados en los Dataframes Static TOTAL (post) es: " + str(sum))
```
